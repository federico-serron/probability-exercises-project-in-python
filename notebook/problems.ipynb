### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import random

#Funtion to roll the dice.
def roll_dices(repeat=1):
    return [np.random.randint(1,7,2) for x in range(repeat)]

# Sum of results of each rolled
total = [np.sum(roll_dices()) for i in range(1,1000+1)]

# Checking out times the results where > 7 or even
even_or_plus_seven = [n for n in total if n % 2 == 0 or n > 7 ]

# Getting probabilty P(>7 | even) = 
p = len(even_or_plus_seven)/len(total)
print('The probability to get an even number or higher than 7 is:','{:,.1%}'.format(p))
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import numpy as np
import random

d = ['white']*10+['red']*20+['green']*30
  
#Initialize important variables
n_simulations = 10000
caseA = 0
caseB = 0

for i in range(n_simulations):
    
    #make a numpy array of the colors that we choose
    # chosen = np.array([d[random.randint(0,59)] for c in range(5)])
    chosen = np.random.choice(d, 5)

    #find the number of each taht we picked
    white = np.sum(chosen == 'white')
    red = np.sum(chosen == 'red')
    green = np.sum(chosen == 'green')
    
    #Keep track if the combination met the criteria
    if white == 3 and red == 2:
        caseA += 1
    elif white == 5 or red == 5 or green == 5:
        caseB += 1
    
#print 'the probability of 3 white and 2 red is' in percentage
print('The probability of 3 white balls and 2 red is', '{:,.1%}'.format(caseA/n_simulations))

#print 'the probability of all the same color is' in percentage   
print('The probability of all the balls same color is', '{:,.1%}'.format(caseB/n_simulations))
```
